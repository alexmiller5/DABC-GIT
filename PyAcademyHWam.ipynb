```python
# Dependencies and Setup
import os
import csv
import pandas as pd
import numpy as np

# set path and load files
Schoolpath = os.path.join("Resources", "schools_complete.csv")
Studentpath = os.path.join("Resources", "students_complete.csv")


school_file = pd.read_csv(Schoolpath)
student_file = pd.read_csv(Studentpath)

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.DataFrame(school_file)
student_df = pd.DataFrame(student_file)

school_df.head(3)

```

```python
student_df.head(3)
```

```python
# Combine the data into a single dataset
district_data = pd.merge(student_file, school_file, how="left", on=["school_name"])
district_df = pd.DataFrame(district_data)

district_df.head(3)
```

```python
#Create single row DF for District_totals: Schools, students, budget, avg math score, avg reading score, %passing math, %passing reading, overall passing rate (avg of % M&R)
    #SCHOOL INFO:
total_schools = len(school_df["school_name"].unique())
total_budget = school_df["budget"].sum()
    
    #STUDENT INFO: avg math score, avg reading score, %passing math, %passing reading, overall passing rate (avg of % M&R)?
total_students = district_df["student_name"].count()
avg_math = district_df["math_score"].mean()
avg_reading = district_df["reading_score"].mean()
total_students = district_df["student_name"].count()
avg_math = district_df["math_score"].mean()
avg_reading = district_df["reading_score"].mean()
pass_math = (district_df['math_score']>=60).sum()/total_students*100
pass_read = (district_df['reading_score']>=60).sum()/total_students*100
pass_rate_overall = (pass_math+pass_read)/2

#DATA FRAME
Dist_overview_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Budget":[total_budget],
    "Total Students":[total_students],
    "Avg Math Score":[avg_math],
    "Avg Reading Score":[avg_reading],
    "Percent Passing Math": [pass_math],
    "Percent Passing Reading": [pass_read],
    "Overall Pass Percentange":[pass_rate_overall]
})
Dist_overview_df
```

```python
#Schools Summary
```

```python
school_type = school_df.set_index(["school_name"])["type"]
school_type
```

```python
#merge both the dataframes
combined_df = pd.merge(student_df,school_df,on="school_name")


#total students per school
total_students_per_school = combined_df["school_name"].value_counts()
```

```python
#total school budget
budget_per_school = combined_df.groupby(["school_name"]).mean()["budget"]

```

```python
#budget per student 
budget_per_student = budget_per_school/total_students_per_school
budget_per_student
```

```python
#Average math Score
#total math score per school
avg_math_scores = round(combined_df.groupby(["school_name"]).mean()["math_score"],2)


```

```python
#average reading scores per school
#total reading scores per school
avg_reading_scores = round(combined_df.groupby(["school_name"]).mean()["reading_score"],2)
```

```python
#students passing math by schools
students_passing_math = combined_df[(combined_df["math_score"]>60)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]
percentage_passing_math_schools = round((students_passing_math_per_school/total_students_per_school)*100,2)
```

```python
#finding % of students passing reading schoolwise

#find the total students passing reading from the combined dataset
students_passing_reading = combined_df[combined_df["reading_score"]>60]

#find the students passing reading schoolwise by doing groupby
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]

#percentage of students passing reading per school
percentage_passing_reading_schoolwise = round((students_passing_reading_per_school/total_students_per_school)*100,2)
```

```python
#overall passing rate per school
overall_passing_rate_schoolwise = (percentage_passing_math_schools + percentage_passing_reading_schoolwise)/2
```

```python
#creating a summary of the school data
school_summary_df = pd.DataFrame({"School Type":school_type,
                                 "Total Students":total_students_per_school,
                                 "Total School Budget":budget_per_school,
                                  "Budget per student":budget_per_student,
                                 "Average Math Score Schoolwise":avg_math_scores,
                                 "Average Reading Score Schoolwise":avg_reading_scores,
                                 "% Passing Math Schoolwise":percentage_passing_math_schools,
                                 "% Passing Reading Schoolwise":percentage_passing_reading_schoolwise,
                                 "Overall Passing rate":overall_passing_rate_schoolwise})
school_summary_df
```
